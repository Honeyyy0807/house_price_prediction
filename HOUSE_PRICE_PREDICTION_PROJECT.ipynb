```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from tkinter import *
from tkinter import messagebox

#Load the dataset
dataset = pd.read_csv('USA_Housing_correct.csv')
# Preprocess the dataset
x=dataset.drop(['Price (In Rupees)','Area of The Plot (In Squarefeet).1'] ,axis=1).values
y = dataset.iloc[:, -1].values

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
# Train the model
regressor = LinearRegression()
regressor
regressor.fit(X_train, y_train)

# Create a GUI
root = Tk()
root.title("House Price Prediction")
root.state("zoomed")
root.config(background='Blue')

root.geometry("400x200")
label = Label(root, text="HOUSE PRICE PREDICTION PROJECT")
label.config(background='Blue',foreground='white', font= ('Times New Roman',30,'bold'))
label.place(relx=0.2,rely=0.1)


# Create labels and entry widgets for input variables
label1 = Label(root, text="AREA OF THE HOUSE (IN SQUARE FEET):")
label1.config(background='Blue',foreground='white', font= ('Times New Roman',15,'bold'))
#label1.grid(row=0, column=0)
label1.place(relx=0.17,rely=0.3)


entry1 = Entry(root)
#entry1.grid(row=0, column=1)
entry1.place(relx=0.5,rely=0.3)

label2 = Label(root, text="NUMBER OF BEDROOMS:")
label2.config(background='Blue',foreground='white', font= ('Times New Roman',15,'bold'))
#label2.grid(row=1, column=0)
label2.place(relx=0.17,rely=0.4)

entry2 = Entry(root)
#entry2.grid(row=1, column=1)
entry2.place(relx=0.5,rely=0.4)

label3 = Label(root, text="POPULATION OF THE LOCATION:")
label3.config(background='Blue',foreground='white', font= ('Times New Roman',15,'bold'))
#label3.grid(row=2, column=0)
label3.place(relx=0.17,rely=0.5)

entry3 = Entry(root)
#entry3.grid(row=2, column=1)
entry3.place(relx=0.5,rely=0.5)

entry4 = Entry(root)
#entry4.grid(row=3, column=1)
entry4.place(relx=0.5,rely=0.6)

label4 = Label(root, text="AGE OF THE HOUSE :")
label4.config(background='Blue',foreground='white', font= ('Times New Roman',15,'bold'))
#label4.grid(row=3, column=0)
label4.place(relx=0.17,rely=0.6)


# Create a function to predict the house price
def predict_price():
    
    size = float(entry1.get())
    bedrooms = int(entry2.get())
    population=int(entry3.get())
    age = int(entry4.get())
    if(population >30000 or population<1000):
        m=messagebox.showerror('error', "population should be greater than 1000 and less than 30000")
    elif (size<272 or size>8160):
        m=messagebox.showerror('error', "size should be greater than 272(1 marla) and less than 8160(30 marla)")
    elif (bedrooms<1 or bedrooms >10):
        m=messagebox.showerror('error', "bedrooms should be greater than 2 and less than equal to 10")
    elif (age<0 or age>30):
        m=messagebox.showerror('error', "age should be greater than 0 and less than equal to 30")
    else:
    # Preprocess the input drata
        input_data = np.array([[size,age,bedrooms,population]])

        
# Make the prediction
        prediction = regressor.predict(input_data)
        
        root.destroy()
        roo=Tk()
        roo.title("House Price Prediction")
        roo.state("zoomed")
        roo.config(background='Blue')
                # Display the prediction
        label4 = Label(roo, text="PREDICTED HOUSE PRICE IS: Rs %.f" % prediction )
        label4.config(background='Blue',foreground='white', font= ('Times New Roman',30,'bold'))
        label4.place(relx=0.23,rely=0.4)
        roo.mainloop()

            # Create a button to make the prediction
button1 = Button(root, text="Predict", command=predict_price)
button1.place(relx=0.45,rely=0.8)

root.mainloop()
```
